### Steganography Tools
*** 

OpenStego is a steganography application with two main functionalities. 

The first is data hiding to hide data within an image file, so it modifies the image to embed data in the image but it also has a watermarking feature to watermark image files to detect unauthorized copying.

If an image file is watermarked with a unique image or a unique signature, it makes imperceptible changes to the image in a unique way, so if a photograph or digital art is sold going against the copyright, then that unique watermark can be used to track down the source of that unauthorized copying. 

Steghide is another steganography application, and it hides confidential files inside an image or audio file with a passphrase and this supports bitmap, BMP images, JPEG image formats, as well as AU and WAV audio files, so in order to extract the secret message or confidential message, the decoder requires a passphrase to be able to extract it successfully.

SNOW is a free steganography application - hides messages inside of ASCII text - is a command line based tool and how this works is that it takes advantage of white space and hides invisible text inside, so it encodes the text inside of the white space, so SNOW will add white space to the file, and it's used in conjunction with the secret key to convert the white space back into the actual message. 

The only difference between the original ASCII text file and the encoded file, the one with the encoded message, is that the encoded version contains a lot of extra white space to hide the message inside of it.

Coagula and Sonic Visualizer are tools used to encode and decode images embedded in audio files.

So the first half of this is Coagula - is a graphical music editing tool - constructs sounds by analyzing the sinewaves within images and then encodes the image by creating an audio file in ``.wav`` format that encodes the image into the audio. 

Then Sonic Visualizer, which is a visualization, analysis, and annotation of audio recordings tool, and this can take a message created with Coagula and decode it so that you can re-extract the image that was encoded in the audio file.

Another class of steganography tool or often used alongside of these tools is TinEye. 

It's an image identification site that tracks where and how images appear online, and you can verify images by finding out where they're appearing on the Internet, so if you upload an image to TinEye, it will display where this image appears on different sites online.

TinEye searches by image or reverse image search, so you can search for images or perform a reverse search to find images that are like the one you're looking for.

Metagoofil is an image gathering tool, an OSINT or open-source intelligence tool. 

This extracts metadata of public documents that belong to a target company. 

It queries in Google for different file types, then downloads to disk for metadata for extraction and report generation.

This is another Kali tool that's installed in addition to the default Kali OS install.

Online SSL checkers are developed for verifying SSL certificates. 

Verifies the expiry date and certificate configuration to check that it's a good configuration. 

It checks the details in the SSL certificate, including the versions, and it verifies domain name mismatches, signature algorithm strength, and the chain of trust. 

- Do you need to use these online SSL checkers?

No. They can be convenient, but as a pentester, you should have all the tools you need installed locally. Using a web browser, OpenSSL, Curl, and nmap, these should be enough to check SSL certificates. Although when sharing results with others, especially including in reports, sometimes the online SSL checkers have a more nicely formatted output that is convenient to use. 

Now, in terms of terminology, when talking about SSL, we're really talking about TLS for the most part. As of today, the strictly SSL variant is now deprecated and the modern TLS, transport layer security, should be used and TLS has three more recent versions, 1.1, 1.2 and 1.3. 

- 1.1 is now officially deprecated and 1.2 is on its way out, so all new TLS certificates should be issued with TLS version 1.3
