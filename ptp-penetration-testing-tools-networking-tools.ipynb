### Networking Tools
*** 

Providing an overview of networking tools, Wireshark and Hping. 

Wireshark is the graphical network protocol analyzer - comes preinstalled in Kali Linux, and it captures packets from a specified connection, so when you run it, you specify the interface you want to capture packets from, and if it's going to capture everything, it needs to be running in a privileged mode using sudo, so that it has permissions to capture everything from the device. 

It's likely the most often used packet sniffer and it's compatible with Windows, Linux, and Mac. 

It can perform several functions, packet capturing, filtering, and visualization of what you're seeing coming over the network.

When you first start Wireshark, you specify the interface and you also specify the protocol you want to capture. 

In Wireshark, there are two types of filtering, capture filtering and display filtering. 

- Capture filtering filters it at the capture level, so when it's capturing from the interface, it discards what it doesn't capture, so it doesn't save it, it doesn't display it in its interface. 


- Display filtering is something you can turn on and off, so it's, only affects what you see, not what Wireshark actually captures and it works really well with other tools, such as tcpdump.

Have Wireshark, the graphical tool on your local system, on your workstation, but then if you run tcpdump on different devices or different servers remotely, you can have tcpdump save its captures as pcap files, packet capture files, that are compatible with Wireshark, so you transfer them to your workstation and load the files in Wireshark so that you can analyze them locally

Some Wireshark use cases are to identify network abuse within a company, looking for particular types of packets or from particular sources or destinations. 

Can monitor web access, looking at only web traffic on your network.

Search for filter specific protocols. If you suspect a particular device is uploading a lot of content over a torrent protocol and it shouldn't be, you can filter out for that specific protocol and look at the devices that you suspect that are transferring this way.

It can be used as sort of a scanner for endpoints, looking at particular protocols - let's say, you suspect a mail server is running an open relay, you can scan the endpoints looking at particular ports and IP addresses, looking for any protocol activity based on that information, and it does it in a graphical way. This might have some functionality overlap with other tools such as nmap, where we might use Wireshark for monitoring packet traffic and use nmap for scanning endpoints. 

Where this really comes into play, is if you scan for nmap and you can't determine if there's any activity on a particular port and maybe the port is open, but it's not active, there's not much you can tell about it, unless you know what's being transmitted - if you have Wireshark monitor a particular host and port, you can see what traffic is going over that. 

Can use Wireshark as sort of a pattern recognition tool to look for patterns in your traffic, to see what is happening, how traffic behaves. In very big networks, on very busy networks, it might be hard to follow everything that's going over your network device, over your interface, which is why capture filtering and display filtering are so important, so you only get what you're looking for, and if you're looking for patterns in particular protocols, Wireshark is a great tool for that.

Hping or Hping version 3 is a GPL software, so it's an open-source software, and it's a command-line-oriented tool for packet assembly and analysis. 

It supports several protocols such as TCP, UDP, ICMP, and raw IP packets.

Can think of it as an advanced ping command, but you can perform more advanced firewall testing, scanning for ports, performing traceroutes, operating system fingerprinting, auditing TCP/IP stacks directly and it's a great tool for learning protocols such as TCP/IP, because you can expect them at a very fine-grained level.

When using Hping or Hping3, you can use it to test networks and firewalls, perform port scanning, host fingerprinting, and you can actually estimate or use it to guess remote uptime and perform an advanced traceroute.

Hping has some distinct tools, but there's a lot of overlap with what you can do with other tools like Wireshark and nmap together. 

One of the neat things about Hping is that it provides scripting through an interactive shell, so you can use the Tcl language, T-c-l, to script commands and script features through Hping and this programmability through Tcl is what gives Hping its distinctiveness and its power, and why you might choose it over other tools that have similar features.
