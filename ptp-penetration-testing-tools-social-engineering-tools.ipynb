### Social Engineering Tools
*** 

Providing an overview of the social engineering tools, namely the SET. the social-engineering toolkit, and BeEF, the browser exploitation framework.

SET, the social-engineering toolkit, is an open source framework designed for social engineering.

It's to perform advanced attacks against the human element in an organization. 

It's Python-driven

Comes preinstalled in Kali OS. 

Often used in conjunction with Metasploit to carry out social-engineering attacks. 

The main benefit of the social-engineering toolkit is that it makes social engineering less time-consuming, so it allows you to develop attacks and test them much quicker than having to do it yourself.

So using SET, SET can be used for multiple attack types, including mass mailers, so sending out mass emails, with phishing attacks included, carrying out spear fishing, targeted phishing attacks, attacking WAPs or wireless access points, performing PowerShell exploits, and web site-based exploits. 

Coming up with clever social-engineering attacks is time-consuming, so having a tool to help to test the variety of attacks without the time and costs to come up with them is a huge saving for pentesters. 

The browser exploitation framework or BeEF is a tool focused on web browsers to test the vulnerability and exposure of the browser to client-side attacks. 

Uses client-side attack vectors, testing the exploitability of a web app's client-side implementation.

This is another tool that comes preinstalled in the Kali Linux operating system.

This tool attempts to answer the question:

- Can sensitive data be taken directly through the browser? 


- Is the client and website they're accessing vulnerable to these attacks?

BeEF comes with a hook - the BeEF hook, a JavaScript file used to hook, exploit, and act as a C&C, a command and control, between target and the attacker.

It gathers information about the target and when hooked, additional functions are available, so using the hook, it's including the JavaScript in an HTML page, using the URL ``http://`` and then some local URL and a given port, in this example, ``192.168.1.15:3000``, so port ``3000/hook.js``. 

The critical part is injecting the hook.js - it's injecting a JavaScript payload into the browser to gather information from the browser, even employing tools like key logging, mouse tracking, playing sounds, gathering cookie information, inspecting credentials that come up on the screen and accessing things like webcam and microphone if they're not secured. 

Testing for these vulnerabilities is important to test how browser environments are configured in an organization and how users respond when these exploits are installed.

In order to get up and running in BeEF, we start the Kali Linux OS and navigate to the BeEF folder. 

Then from the command line, we start the Apache server and modify the default test page used. To verify that we can use BeEF, we modify the test page or an HTML page of our own, and we include the BeEF hook, so we reference the script source to the hook.js wherever we have that running from.

Then we open the BeEF GUI, the web-based graphical user interface, and navigate to its Commands tab, and there we can test the browser security using this framework.
