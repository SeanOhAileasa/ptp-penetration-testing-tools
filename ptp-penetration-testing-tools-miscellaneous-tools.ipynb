### Miscellaneous Tools
*** 

Presenting some comparisons on miscellaneous tools available to penetration testers.

SearchPloit is a command line search tool for the Exploit-DB. 

It copies the Exploit database locally. 

Useful for security assessments, for segregated networks with no Internet access, or for the convenience of having the exploit db offline instead of going through the web interface at [ [exploit-db.com](https://exploit-db.com) ].

This is installed by Kali Linux by default. 

PowerSploit is a collection of Microsoft PowerShell module exploits. 

This aids testers during all phases of an assessment using DLL injection, shell code, WMI commands with persistence options, antivirus bypass, data exfiltration, privilege escalation, and other methods that take advantage of the PowerShell environment. 

This is also installed in Kali Linux by default.

Responder and Impacket are network-based penetration testing tools.

Responder is a network discovery and intercept tool. 

It poisons LLMNR, NBT-NS, and MDNS, so link-local multicast name resolution, NetBIOS name service and multicast DNS. 

By default, it answers to File Server Service requests, S-M-B or SMB, and targets answers only, which provides more stealth on the network, so it's not broadcasting on its own.

Impacket is a collection of Python scripts. 

Targets Windows network protocols.

Used to enumerate users and groups, capture hashes, it moves laterally, and escalates privileges, and can also include remote execution testing with Psexec and Smbexec. It can look up our PCM points, and this works all from the command line tools of Impacket.

For Empire, we need to talk about it in the past tense.

It was a pure PowerShell post-exploitation agent, built on secure communications and a flexible architecture. It's used to run PowerShell without the need of powershell.exe and it can be used to rapidly deploy modules, however, Empire is now archived or deprecated. 

In general, it's no longer supported. On its GitHub repository, it has a message, ``This repository has been archived by the owner. It is now read-only.`` This is because it's built on Python 2, which is now end-of-life and on other outdated dependencies.

Its usefulness was reached about five or six years ago and hasn't been updated since, however, it may be part of legacy workflows and deployments, so targeting older systems and if a pen testing operation has any virtual machines running older legacy workflows, with Empire already installed, it may still be in use, but it won't be used for new deployments, new pen testing setups.

Starting out fresh today with a pen testing setup would not include Empire in the array of tools that would be used.

Metasploit, on the other hand, is used, it's used a lot. 

It's software from Rapid7, but it's also open source. 

The Metasploit framework is included in the Kali Linux operating system, and it's a tool to probe systematic vulnerabilities on networks and servers.

It comes with an array of payloads with ready-made or custom code that can be configured and tested from Metasploit. 

It's used by DevSecOps and hackers as well. 

Metasploit is the big framework for pen testers, the kitchen sink, and it has everything under the sun. It's the tool of tools, payloads, exploits, miscellaneous tools and tests. 

It's great for checking known exploits, and it has a built-in extensive database of options that can be used

- mitm6 is a pen testing tool that exploits Windows default configurations. 

It can spoof DNS replies and act as a malicious DNS server carrying out a man-in-the-middle attack, and it can redirect traffic to a specified endpoint controlled by the attacker or controlled by the pen tester in the case of testing. 

It's designed to work together with ntlmrelayx from impacket. 

- CrackMapExec is a post-exploitation tool. 

It automates accessing security Active Directory networks and makes heavy use of the impacket library. 

It's intended mainly for offensive security, but obviously used defensively to assess account privileges and identify any misconfigurations. 

As a pen tester, you'll use these tools to test to see if the networks or configurations are susceptible to these tools and the exploits that they convey.
