### Cloud Tools
*** 

Scout Suite is an open source multi-cloud security auditing tool. It provisions security posture assessments in cloud environments. It gathers data for inspection and identifies risky areas and it provides an automated clear view of attack surfaces, and it's written in Python - Scout Suite generates a browsable HTML set of reports and it's used by pentesters to provide a point in time view of a cloud service. 

The cloud providers that Scout Suite covers are the Amazon Web Services, AWS, Microsoft Azure, the Google Cloud Platform, or GCP, and the Alibaba Cloud, and Oracle Cloud.

CloudBrute is a multi-platform tool - runs cross platform on Windows, Linux, and Mac and it searches and enumerates target company's cloud infrastructure, looking at files, open buckets, in terms of Amazon Web Services, applications, databases, and it covers a similar list of cloud providers targeting Amazon Web Services, AWS, Microsoft Azure, Google Cloud Platform, Alibaba Cloud, Vultr, Digital Ocean, and Linode. 

Now remember, it's very important to get permission before you carry out any pen testing against these cloud providers.

Make sure your organization knows and that you interface with a cloud provider before carrying out these tests. 

The primary use of tools like Scout Suite and CloudBrute will be to find misconfigurations, so if something is misconfigured, it will lead to data breaches, usually by exposing more data than necessary because of a misconfigured access control. 

Some organizations may not even be aware that it's using some of these cloud services - sometimes it's done through third-parties, third-party abstraction, leading to an unexpected dependency with the cloud provider or an employee implementing a new feature or a new tool using a third-party cloud without approval. 

CloudBrute helps find these dependencies as well.

Pacu is an open source, A-W-S or AWS exploitation framework. It's designed for testing against cloud environments through AWS security and it exploits configuration flaws using modules, modules to check permissions, privilege separation, weak credentials or open credentials, or remote code execution in the cloud environment and it's written in Python 3.

Pacu has been described as the Nessus for the AWS cloud. 

Cloud Custodian is configured using a rules engine using YAML DSL, a domain-specific language to describe its rules.

It manages public cloud accounts and resources.

Allows users to define policies for managing cloud infrastructure.

Contains a library of actions and filters to build policies.

Cloud Custodian supports A-W-S or AWS, Amazon Web Services, Microsoft Azure, and the Google Cloud Platform, GCP.

It can help enforce policies with metrics and structured outputs and it gives you a detailed report of the cloud infrastructure you're analyzing. 

It could also help you clean up unused resources and carry out cost management based on the resources you're using. 

Can also validate using dry runs and review changes to your cloud policies. These policies are expressed in the YAML rules engine, where you specify the type of resource, the filters, and the actions to take.
